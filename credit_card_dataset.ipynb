```python
check_missing_values = df.isna().sum()
print(check_missing_values)
print("there are no missing or null values in this data frame.")
```

```python
transection_count = df['Class'].value_counts()
print('There are ', transection_count[0],'legitimate transections')
print('There are ', transection_count[1],'fraudulent transections')
```

```python
#What percentage of transactions are fraudulent? using
percentage_fraudulent_transections = df['Class'].value_counts(normalize=True)[1] *100
print(f'{percentage_fraudulent_transections:.2f}% of the transactions are fraudulent.')
```

```python
min=df['Amount'].min()
max=df['Amount'].max()
mean=df['Amount'].mean()
median=df['Amount'].median()
print(f'The minimum amount is {min}')
print(f'The maximum amount is {max}')
print(f'The mean amount is {mean}')
print(f'The median amount is {median}')
```

```python
import matplotlib.pyplot as plt

transection_count = df['Class'].value_counts()
plt.bar(transection_count.index, transection_count.values , color=['lightblue' , 'orange'],edgecolor='black')
plt.xticks([0, 1], ['Legitimate', 'Fraudulent'])
plt.xlabel('Transaction Type')
plt.ylabel('Transaction Count')
plt.title('Count of Fraudulent vs Legitimate Transactions')
plt.show()
```

```python

df['Amount']
plt.hist(df['Amount'], bins=30, color='lightpink' , edgecolor='black')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Histogram of Transaction Amounts')
plt.show()
```

```python
import plotly.express as px
fig =px.pie(df, names='Class', title='Distribution of Fraudulent and Legitimate Transactions')
fig.show()
```

```python
import seaborn as sns
correlation_matrix = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="viridis", cbar=True)

plt.title("Correlation Btw Numerical Features")
plt.xlabel("Features")
plt.ylabel("Features")
plt.show()
```
